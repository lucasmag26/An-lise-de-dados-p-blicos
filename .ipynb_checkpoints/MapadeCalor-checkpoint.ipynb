```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
import matplotlib.ticker as ticker
from datetime import datetime
import folium
from folium import plugins
from folium.plugins import HeatMap
%matplotlib inline
```

```python
dados_roubo = pd.read_csv(r'C:\Users\gabri\Lucas\DadosBO_2020(ROUBO DE VEÍCULOS).csv', low_memory=False,delimiter=';', encoding='iso-8859-1', decimal=',')
dados_roubo.info()
```

```python
dados_roubo.rename(columns={'DATAOCORRENCIA': 'DATA_DE_OCORRENCIA','DESCR_MARCA_VEICULO': 'MARCA_DO_VEICULO',
                            'PERIDOOCORRENCIA':'PERIODO_DE_OCORRENCIA','DATAELABORACAO': 'DATA_DE_ELABORACAO'}, inplace=True)
dados_roubo.sample(10)
```

```python
mapa = folium.Map([-23.218113, -45.893545], zoom_start=11)
dados_roubo=dados_roubo.dropna(subset=['LATITUDE'])
dados_roubo=dados_roubo.dropna(subset=['LONGITUDE'])
local = dados_roubo[['LATITUDE', 'LONGITUDE']].values
HeatMap(local).add_to(mapa)
mapa
```

```python
dados_roubo['BAIRRO'].value_counts().head(20)
```

```python
dados_roubo.groupby('DESCR_COR_VEICULO').size().sort_values().tail(5).plot(kind='bar', figsize=(15,10), fontsize = 'medium')
plt.title('CORES DE VEÍCULOS MAIS ROUBADOS - SJC (2020)')
plt.ylabel('QUANTIDADE')
plt.xlabel('CORES')
plt.grid(True, which='major')
```

```python
dados_roubo.groupby('DESCR_TIPO_VEICULO').size().sort_values().tail(5).plot(kind='pie', figsize=(15,10), fontsize = 'medium')
plt.title('TIPO DE VEÍCULOS MAIS ROUBADOS - SJC (2020)')
plt.grid(True, which='major')
```

```python
dados_roubo.groupby('MARCA_DO_VEICULO').size().sort_values().tail(10).plot(kind='bar', figsize=(15,10), fontsize = 'medium')
plt.title('TIPO DE VEÍCULOS MAIS ROUBADOS - SJC (2020)')
plt.ylabel('QUANTIDADE')
plt.xlabel('TIPO DE VEÍCULO')
plt.grid(True, which='major')
```
